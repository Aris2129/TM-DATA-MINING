```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

df = pd.read_csv("Mall_Customers.csv")

print("===== DATA AWAL =====")
print(df.head())

print("\n===== INFO DATASET =====")
print(df.info())

print("\n===== STATISTIK DESKRIPTIF =====")
print(df.describe())

print("\nJumlah Record :", df.shape[0])
print("Jumlah Atribut :", df.shape[1])

print("\n===== CEK MISSING VALUE =====")
print(df.isnull().sum())


le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

X = df[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

print("\n===== DATA SEBELUM SCALING =====")
print(X.head())

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\n===== DATA SESUDAH SCALING =====")
print(pd.DataFrame(X_scaled, columns=X.columns).head())

inertia = []
range_k = range(1, 11)

for k in range_k:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range_k, inertia, marker='o')
plt.title("Elbow Method untuk Menentukan Jumlah Cluster Optimal")
plt.xlabel("Jumlah Cluster (K)")
plt.ylabel("Inertia")
plt.grid()
plt.show()

kmeans_final = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans_final.fit_predict(X_scaled)

sil_score = silhouette_score(X_scaled, df['Cluster'])
print("\nSilhouette Score :", round(sil_score, 3))

print("\n===== RATA-RATA TIAP CLUSTER =====")
print(df.groupby('Cluster').mean())

plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='Cluster',
    palette='tab10',
    data=df
)
plt.title("Hasil Segmentasi Pelanggan Mall")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1100)")
plt.legend(title="Cluster")
plt.show()

print("\nKESIMPULAN:")
print("1. Dataset berhasil dikelompokkan menjadi 5 cluster pelanggan.")
print("2. Segmentasi didasarkan pada usia, pendapatan, dan perilaku belanja.")
print("3. Silhouette Score menunjukkan kualitas clustering yang cukup baik.")
```
